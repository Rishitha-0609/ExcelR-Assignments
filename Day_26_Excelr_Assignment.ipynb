Coding Exercise: Generative AI Applications (Text Summarization and Content Generation)

Write a Python program using the Hugging Face 'transformers' library to summarize a short paragraph. Use any open source LLM model from the hub.

```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

text = """Artificial intelligence (AI) is transforming various industries by automating tasks,
improving efficiency, and enabling new capabilities. From healthcare to finance, AI-powered
technologies are being used to analyze data, predict outcomes, and assist in decision-making.
As AI continues to evolve, it is expected to revolutionize the way we work and interact with technology."""

summary = summarizer(text, max_length=50, min_length=20, do_sample=False)

print(summary[0]["summary_text"])
```
